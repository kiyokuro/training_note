````python
from csv import reader

def open_dataset(dataset_path):
    opened_file = open(dataset_path)
    read_file = reader(opened_file)
    return list(read_file)
    
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
        
def deplicate_app(dataset, name_column_index):
    deplicate_apps_with_number = {}
    for app in dataset:
        app_name = app[name_column_index]
        if app_name in deplicate_apps_with_number:
            deplicate_apps_with_number[app_name] += 1
        else:
            deplicate_apps_with_number[app_name] = 1
            
    tow_or_more_apps = {}
    for key, value in deplicate_apps_with_number.items():
        if value > 1:
            tow_or_more_apps[key] = value
    print(str(len(tow_or_more_apps)) + ' deplicate apps')
        
    for key, value in tow_or_more_apps.items():
        print(key + ' : ' + str(value))
        
        
ios_data = open_dataset(dataset_path='./datasets/AppleStore.csv')
google_data = open_dataset(dataset_path='./datasets/googleplaystore.csv')
# delete wrong data, https://www.kaggle.com/lava18/google-play-store-apps/discussion/66015
del google_data[10473]
````

````python
print('ios data')
explore_data(ios_data, 0, 1)
explore_data(ios_data, 1, 3, True)
print('google data')
explore_data(google_data, 0, 1)
explore_data(google_data, 1, 3, True)
````

````python
deplicate_apps = []
uniqu_apps = []
for app in google_data[1:] :
    name = app[0]
    if name in uniqu_apps:
        deplicate_apps.append(name)
    else:
        uniqu_apps.append(name)
        
print(len(deplicate_apps))
````

````python
# deplicate apps and header equal 1181 + 1
print('expected num eq ' + str(len(google_data) - 1182))
review_max = {}
for app in google_data[1:]:
    name = app[0]
    n_review = float(app[3])
    if name in review_max and review_max[name] < n_review:
        review_max[name] = n_review
    if name not in review_max:
        review_max[name] = n_review
        
print(len(review_max))
````

````python
# ÈáçË§á„Åó„Å¶„ÅÑ„Çã„Éá„Éº„Çø„ÅØ„É¨„Éì„É•„ÉºÊï∞„Åå„ÇÇ„Å£„Å®„ÇÇÂ§ö„ÅÑÁâ©„ÇíÊé°Áî®„Åô„Çã
android_clean = []
already_added = []

for app in google_data[1:]:
    name = app[0]
    n_review = float(app[3])
    if name in review_max and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
        
print(len(android_clean))        
````

````python
# Filter only english apps
def ensure_english_app(app_name):
    for char in app_name:
        if ord(char) > 127:
            return False
    return True

# function unit test
test_app_name = ['Instagram', 'Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠', 'Docs To Go‚Ñ¢ Free Office Suite', 'Instachat üòú']
for index, app_name in enumerate(test_app_name):
    if ensure_english_app(app_name):
        print(str(index) + ' is english app')
    else:
        print(str(index) + ' is not english app')
    
# build english app list
english_ios = []
english_android = []

for app in ios_data[1:]:
    if ensure_english_app(app[2]):
        english_ios.append(app)
for app in android_clean:
    if ensure_english_app(app[0]):
        english_android.append(app)
        
print(len(english_ios))
print(len(english_android))
````
